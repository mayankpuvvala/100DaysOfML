##remove the columns crime_no, Arr_ID, FIRNo, AccusedName, Person_Name, PresentAddress, Nationality_Name, DOB, Caste, Profession, PresentState, PermanentAddress, PermanentCity, PermanentState,

```python
columns_to_remove = ['crime_no', 'Arr_ID', 'FIRNo', 'AccusedName', 'Person_Name', 
                     'PresentAddress', 'Nationality_Name', 'DOB', 'Caste', 
                     'Profession', 'PresentState', 'PermanentAddress', 
                     'PermanentCity', 'PermanentState']

df.drop(columns=columns_to_remove, inplace=True)
df
```

```python
df.info()
```

```python
df['Year'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder

# Create a label encoder object
label_encoder = LabelEncoder()

# Iterate through each column in the DataFrame
for column in df.columns:
    # Check if the column dtype is object (i.e., categorical)
    if df[column].dtype == 'object':
        # Apply label encoding to the column
        df[column] = label_encoder.fit_transform(df[column])

# Display the DataFrame after label encoding
df
```

```python
# Convert 'Year' and 'Month' columns to int8
df['Year'] = df['Year'].astype('int8')
df['Month'] = df['Month'].astype('int8')

# Remove the 'Person_No' column
if 'Person_No' in df.columns:
    df.drop(columns=['Person_No'], inplace=True)
df.info()
```
